```python
non_crim = df.loc[df['Category'] == 'NON-CRIMINAL']
```

```python
non_crim.shape
```

```python
non_crim50 = non_crim[:50]
```

```python
from folium import plugins

# let's start again with a clean copy of the map of San Francisco
san_fran = folium.Map(location = [latitude, longitude], zoom_start = 12)

# instantiate a mark cluster object for the incidents in the dataframe
incidents = plugins.MarkerCluster().add_to(san_fran)

# loop through the dataframe and add each data point to the mark cluster
for lat, lng, label, in zip(non_crim50.Y, non_crim50.X, non_crim50.Category):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(incidents)

# display map
san_fran
```
